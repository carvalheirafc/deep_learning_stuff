```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
diabetes = pd.read_excel('diabetes_Ex1.xlsx')
```

```
diabetes = diabetes.drop(['Unnamed: 16'], axis=1)
diabetes = diabetes[:-8]
diabetes['C'] = pd.to_numeric(diabetes['C'])
diabetes.info()
```

```
diabetes.describe()
```

```
fig, ax = plt.subplots(17, sharex=True)
for i in range(0, len(list(diabetes))):
  sns.distplot(diabetes[list(diabetes)[i]], ax=ax[i])
```
