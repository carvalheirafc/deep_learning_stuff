```python
import json 
import matplotlib.pyplot as plt
```

```python
alpha = json.load(open('history/alpha', 'r'))
beta = json.load(open('history/beta', 'r'))
victor = json.load(open('history/victor', 'r'))
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax_1 = plt.plot(alpha['loss'], label='Loss: alpha',)
ax_1 = plt.plot(alpha['categorical_accuracy'], label='Acc: alpha')
ax_1 = plt.legend()
ax_2 = plt.plot(beta['loss'], label='Loss: beta')
ax_2 = plt.plot(beta['categorical_accuracy'], label='Acc: beta')
ax_2 = plt.legend()
ax_3 = plt.plot(victor['loss'], label='Loss: victor')
ax_3 = plt.plot(victor['categorical_accuracy'], label='Acc: victor')
ax_3 = plt.legend()

plt.title('Loss and Categorical Accuracy from three LeNet Models')
plt.xlabel('Epochs')
plt.show()
```
